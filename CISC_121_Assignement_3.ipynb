```python
class Solution:
    def searchRange(self, nums: List[int], target: int) -> List[int]:

        def binary_search(nums, target, is_searching_left):
            left = 0
            right = len(nums) - 1
            idx = -1

            while left < right:
                mid = (left + right) // 2

                if nums[mid] > target:
                    right = mid - 1
                elif nums[mid] < target:
                    left = mid + 1
                else:
                    idx = mid
                    if is_searching_left:
                        right = mid - 1
                    else:
                        left = mid + 2

            return idx

        left = binary_search(nums, target, True)
        right = binary_search(nums, target, False)

        return [left, right]
```

When nums[mid] == 1 just increment mid. When nums[mid] == 2 swap with high and decrement high but do not increment mid, because the swapped-in element at mid must be re-checked.

```python
def sortColors(nums):
    low = mid = 0
    high = len(nums)

    while mid <= high:
        if nums[mid] == 0:
            nums[low], nums[mid] = nums[mid], nums[low]
            low += 1
            mid += 1
        elif nums[mid] == 1:
            mid += 1
        else:
            nums[mid], nums[high] = nums[high], nums[mid]
            high -= 1
            mid += 1
```
