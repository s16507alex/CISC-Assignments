```python
# Your draft code here (does not need to be fully correct)
class Solution:
    def reverse(x: int) -> int:
        INT_MAX = 2**31 - 1
        INT_MIN = -2**31
        rev = 0

        while x != 0:
            pop = x % 10

            x = x // 10

            if rev > INT_MAX // 10 or rev < INT_MIN // 10:
                return 0

            rev = rev * 10 + pop

        return rev
```

```python
# Final working solution
class Solution:
    def reverse(x: int) -> int:
        INT_MAX = 2**31 - 1
        INT_MIN = -2**31
        rev = 0

        while x != 0:
            pop = x % 10 if x > 0 else x % -10
            x = int(x / 10)

            if rev > INT_MAX // 10 or (rev == INT_MAX // 10 and pop > 7):
                return 0
            if rev < INT_MIN // 10 or (rev == INT_MIN // 10 and pop < -8):
                return 0

            rev = rev * 10 + pop

        return rev
```

I learned that using a two-pointer approach is memory-efficient. The key idea is to move two pointers from opposite ends of the string toward the center, skipping any non-alphanumeric characters and comparing letters in lowercase.This avoids creating a new string and reduces extra space usage to O(1) while keeping the time complexity at O(n).

```python
# Your draft code here (does not need to be fully correct)
class Solution:
    def isPalindrome(self, s: str) -> bool:
        left, right = 0, len(s) - 1

        while left < right:
            while not s[left].isalnum():
                left += 1
            while not s[right].isalnum():
                right -= 1

            if s[left] != s[right]:
                return False

            left += 1
            right -= 1

        return True
```

```python
# Final working solution
# [AI Disclaimer example: Used ChatGPT with prompt ‚ÄúHow to fix boundary checks in palindrome two-pointer code‚Äù only for debugging and concept]
class Solution:
    def isPalindrome(self, s: str) -> bool:
        left, right = 0, len(s) - 1

        while left < right:
            while left < right and not s[left].isalnum():
                left += 1
            while left < right and not s[right].isalnum():
                right -= 1

            if s[left].lower() != s[right].lower():
                return False

            left += 1
            right -= 1

        return True
```
